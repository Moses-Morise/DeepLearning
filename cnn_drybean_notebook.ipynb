```python
# Erster Run: Basismodell MobileNetV2 mit grundlegenden Schichten und 30 Epochen Training, 20 Epochen Fine-Tuning
# Importiere notwendige Bibliotheken
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

# Konfiguriere TensorFlow, um GPU 0 und 1 zu verwenden
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        # Beschrnke TensorFlow darauf, nur die ersten zwei GPUs zu verwenden
        tf.config.experimental.set_visible_devices(gpus[:2], 'GPU')
        
        # Setze das Speicherwachstum, um zu verhindern, dass TensorFlow den gesamten GPU-Speicher von Anfang an belegt
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
    except RuntimeError as e:
        # Sichtbare Gerte mssen festgelegt werden, bevor die GPUs initialisiert werden
        print(e)

# Lade den Datensatz 
data = tf.keras.utils.image_dataset_from_directory(
    'data',                # Verzeichnis, das die Bilder enthlt
    image_size=(600, 600), # Gre der Bilder, die geladen werden sollen
    batch_size=8,          # Anzahl der Bilder pro Batch
    label_mode='categorical' # Kategorische Labels
)

# Erstelle einen Dateniterator
data_iterator = data.as_numpy_iterator()

# Hole einen Batch von Daten
batch = data_iterator.next()

# Zeige die Formen der Daten im Batch an
print(batch[0].shape)

# Visualisiere die Bilder im Batch
fig, ax = plt.subplots(ncols=4, figsize=(20, 20))
for idx, img in enumerate(batch[0][:4]):
    ax[idx].imshow(img.astype(int))     # Zeige das Bild an
    ax[idx].title.set_text(batch[1][idx]) # Setze den Titel als das Label des Bildes
plt.savefig('batch_visualization.png')  # Speichere die Visualisierung

# Teile die Daten auf
train_size = int(len(data) * .7)  # 70% der Daten fr das Training
val_size = int(len(data) * .2)    # 20% der Daten fr die Validierung
test_size = int(len(data) * .1)   # 10% der Daten fr das Testen

# Zuweise die Datenstze
train = data.take(train_size)  # Trainingsdatensatz
val = data.skip(train_size).take(val_size)  # Validierungsdatensatz
test = data.skip(train_size + val_size).take(test_size)  # Testdatensatz

# Baue das Modell
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

model = Sequential()
# Erste Convolutional Layer
model.add(Conv2D(filters=64, kernel_size=(3, 3), padding='same', activation='relu', input_shape=(600, 600, 3)))
model.add(MaxPooling2D(pool_size=(2, 2), strides=2))
# Zweite Convolutional Layer
model.add(Conv2D(filters=128, kernel_size=(3, 3), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2), strides=2))
# Dritte Convolutional Layer
model.add(Conv2D(filters=256, kernel_size=(3, 3), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2), strides=2))
# Vierte Convolutional Layer
model.add(Conv2D(filters=512, kernel_size=(3, 3), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2), strides=2))
# Flatten Layer, um in Dense Layer berzugehen
model.add(Flatten())
# Erste Dense Layer
model.add(Dense(units=512, activation='relu'))
# Dropout Layer, um Overfitting zu verhindern
model.add(Dropout(0.5))
# Ausgangsschicht, die die Klassen vorhersagt
model.add(Dense(units=7, activation='softmax'))  # Angepasst auf 7 Einheiten fr 7 Klassen

# Kompiliere das Modell
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Zeige die Modellzusammenfassung
model.summary()

# Trainiere das Modell
logdir = 'logs'  # Verzeichnis fr TensorBoard Logs
tensor_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
hist = model.fit(train, epochs=20, validation_data=val, callbacks=[tensor_callback])

# Plot Funktion fr Trainingsleistung
def plot_history(history):
    fig, axs = plt.subplots(2, 1, figsize=(10, 10))

    # Plot der Trainings- und Validierungsgenauigkeit
    axs[0].plot(history.history['accuracy'])
    axs[0].plot(history.history['val_accuracy'])
    axs[0].set_title('Model accuracy')
    axs[0].set_ylabel('Accuracy')
    axs[0].set_xlabel('Epoch')
    axs[0].legend(['Train', 'Validation'], loc='upper left')

    # Plot des Trainings- und Validierungsverlustes
    axs[1].plot(history.history['loss'])
    axs[1].plot(history.history['val_loss'])
    axs[1].set_title('Model loss')
    axs[1].set_ylabel('Loss')
    axs[1].set_xlabel('Epoch')
    axs[1].legend(['Train', 'Validation'], loc='upper left')

    # Speichere die Plots
    plt.tight_layout()
    plt.savefig('training_performance.png')

# Rufe die Plot Funktion auf
plot_history(hist)

# Speichere das Modell
model.save('model.h5')
```
